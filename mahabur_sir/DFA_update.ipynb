```python
from automata.fa.dfa import DFA

from visual_automata.fa.dfa import VisualDFA
```

```python
dfa = DFA(
    states={'q0', 'q1', 'q2', 'q3'},
    input_symbols={'0', '1'},
    transitions={
        'q0': {'0': 'q1', '1': 'q2'},
        'q1': {'0': 'q0', '1': 'q3'},
        'q2': {'0': 'q3', '1': 'q0'},
        'q3': {'0': 'q2', '1': 'q1'}
    },
    initial_state='q0',
    final_states={'q0'}
)
```

```python
dfa = VisualDFA(dfa)
dfa # transitions

dfa.table
```

```python
dfa.show_diagram()
dfa.show_diagram(view=True)
```
